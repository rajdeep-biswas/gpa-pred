```python
import pandas as pd
```

```python
df = pd.read_excel('gpapred.xlsx', engine='openpyxl')
```

```python
df
```

```python
import numpy as np
```

```python
score_X_train = []
```

```python
score_X_train.append(df[["Midsem", "Comp"]][0:4]) # , "Credits"
```

```python
score_X_train.append(df[["Midsem", "Comp"]][5:9])
```

```python
score_X_test = df[["Midsem", "Comp"]][10:14]
```

```python
score_X_train = np.array(score_X_train)
```

```python
score_X_train
```

```python
score_X_train.shape
```

```python
score_X_test = np.array(score_X_test)
```

```python
score_X_test
```

```python
score_X_test.shape
```

```python
score_y_train = []
```

```python
score_y_train.append(df[["Grade"]][0:4])
```

```python
score_y_train.append(df[["Grade"]][5:9])
```

```python
score_y_train = np.array(score_y_train)
```

```python
score_y_train
```

```python
score_y_train.shape
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model_scores = Sequential([
    Dense(12, activation='relu', input_shape=(score_X_train.shape[1], score_X_train.shape[2])),
    Dense(48, activation='relu'),
    Dense(36, activation='relu'),
    Dense(score_y_train.shape[2])
])
```

```python
model_scores.summary()
```

```python
model_scores.compile(optimizer='adam', loss='mse', metrics=['mae', 'accuracy'])
```

```python
history_scores = model_scores.fit(score_X_train, score_y_train, epochs=10000, verbose=2)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(history_scores.history['accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

plt.plot(history_scores.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
scores_y_pred = model_scores.predict(score_X_test)
```

```python
scores_y_pred
```
